```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
nl = pd.read_excel('graph.xlsx', sheet_name= '6nl')
nzo = pd.read_excel('graph.xlsx', sheet_name= 'NZO')
apo = pd.read_excel('graph.xlsx', sheet_name= 'APO')
nl = nl[['PC1', 'PC2']]
scaled_nl = StandardScaler().fit_transform(nl)

nzo = nzo[['PC1', 'PC2']]
scaled_nzo = StandardScaler().fit_transform(nzo)

apo = apo[['PC1', 'PC2']]
scaled_apo = StandardScaler().fit_transform(apo)
```

```python
#initialize kmeans parameters
kmeans_kwargs = {
"init": "random",
"n_init": 10,
"random_state": 1,
}

#create list to hold SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_nl)
    sse.append(kmeans.inertia_)

#visualize results
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()

#instantiate the k-means class, using optimal number of clusters
kmeans = KMeans(init="random", n_clusters=3, n_init=10, random_state=1)

#fit k-means algorithm to data
kmeans.fit(scaled_nl)

#view cluster assignments for each observation
kmeans.labels_

#append cluster assingments to original DataFrame
nl['cluster'] = kmeans.labels_

#view updated DataFrame
print(nl)

# array([1, 1, 1, 1, 1, 1, 2, 2, 0, 0, 0, 0, 2, 2, 2, 0, 0, 0])
```
